```python
import requests
from bs4 import BeautifulSoup
url = r"https://www.moh.gov.bh/HealthInfo/DrugDictionary?lang=en"
url1 = r"https://www.moh.gov.bh"
page  = requests.get(url).text
soup= BeautifulSoup(page, "html.parser")
exp = soup.find("div", class_="AtoZ")

Brand_Name = []
href = []
for card in exp.find_all('a'):   
    href.append(url1+card['href'])
    Brand_Name.append(card.text)
```

```python
# creating the dataframe and assigning each column with the data from the lists 

df = pd.DataFrame({
    'Brand_Name':Brand_Name,
    'href':href
    })
df
```

```python
df['med_id'] = df['href'].apply(lambda x:int(str(x).split('/')[-1][:-7]))
df
```

```python
a,b,c,d,e = [],[],[],[],[]

for ids in med_id5:
    url_link = 'https://www.moh.gov.bh/HealthInfo/DrugDetail/{}?Desc=1?lang=en'.format(ids)
    print(med_id.index(ids)+1, ids, url_link)
    driver = webdriver.Chrome(executable_path = Chrome_path)
    
    driver.get(url_link)
    
    driver.find_element_by_xpath('//*[@id="topmenuitem"]/a[6]').click()
    l=driver.find_elements_by_xpath ('//*[@id="renderbody"]/table')

    aa = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[2]/td[2]')
    for i in aa :
        a.append(i.text)
#         print(i.text)
    bb = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[3]/td[2]')
    for i in bb :
        b.append(i.text)
#         print(i.text)
    cc = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[4]/td[2]')
    for i in cc :
        c.append(i.text)
#         print(i.text)
    dd = driver.find_elements_by_xpath('//*[@id="body"]')
    for i in dd :
        d.append(i.text)
    #     print(i.text)
    ee = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[6]/td[2]/ul/li')
    brandName = []
    for i in ee :
        brandName.append(i.text)
    e.append(brandName)
    driver.close ()
    time.sleep(2)
    driver.quit()
    
ddd = [i for i in d if i!='']
data1 = pd.DataFrame({
    'Drug Generic Name': a,
    'Drug Class':b,
    'Chapter':c,
    'Details':ddd,
    'Brand Name':e
    })


data1.to_csv("med_id5.csv", index=False)
data1.head()
```
