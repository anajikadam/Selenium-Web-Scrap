```python
a,b,c,d,e = [],[],[],[],[]

for ids in med_id150:
    url_link = 'https://www.moh.gov.bh/HealthInfo/DrugDetail/{}?Desc=1?lang=en'.format(ids)
    print(med_id150.index(ids)+1, ids, url_link)
    driver = webdriver.Chrome(executable_path = Chrome_path)
    
    driver.get(url_link)
    
    driver.find_element_by_xpath('//*[@id="topmenuitem"]/a[6]').click()
    l=driver.find_elements_by_xpath ('//*[@id="renderbody"]/table')

    aa = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[2]/td[2]')
    for i in aa :
        a.append(i.text)
#         print(i.text)
    bb = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[3]/td[2]')
    for i in bb :
        b.append(i.text)
#         print(i.text)
    cc = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[4]/td[2]')
    for i in cc :
        c.append(i.text)
#         print(i.text)
    dd = driver.find_elements_by_xpath('//*[@id="body"]')
    for i in dd :
        d.append(i.text)
    #     print(i.text)
    ee = driver.find_elements_by_xpath('//*[@id="myTable"]/tr[6]/td[2]/ul/li')
    brandName = []
    for i in ee :
        brandName.append(i.text)
    e.append(brandName)
    driver.close ()
#     time.sleep(2)
    driver.quit()
    
ddd = [i for i in d if i!='']
data1 = pd.DataFrame({
    'Drug Generic Name': a,
    'Drug Class':b,
    'Chapter':c,
    'Details':ddd,
    'Brand Name':e
    })


data1.to_csv("med_id150.csv", index=False)
data1.head()
```

```python
def BrandName1(data):
    Brand_Name1 = []
    for i in data['Brand Name'].tolist():
        if i.count('\'')==2:
            Brand_Name1.append(i.split("'")[1])
        else:
            Brand_Name1.append(i.replace('\'','').replace('[','').replace(']','').split(','))
    data['Brand_Name1'] = Brand_Name1
    return data

final1 = BrandName1(data0)
final2 = BrandName1(data1)
final3 = BrandName1(data2)
final4 = BrandName1(data3)
final5 = BrandName1(data4)
final6 = BrandName1(data5)
final7 = BrandName1(data6)
final8 = BrandName1(data7)
```

```python
# -- coding: utf-8 --

import requests
import re

__base_url = "http://translate.google.com/translate_a/t"
__headers = {
    "User-Agent": "Mozilla/5.0 (Windows; U; en) AppleWebKit/420+ (KHTML, like Gecko) "
                  "Version/3.1.2 Safari/525.21"
}
params = {
    "client": "t",
    "ie": "UTF-8",
    "oe": "UTF-8",
    "sl": "en",
    "tl": "ja",
    "text": "こんにちはモンキー！"
}
matcher = re.compile(r'\[\[\["(.*?)","(.*)","(.*)",""\]\]')


def do_trans(text, tolang, fromlang):
    params['sl'] = fromlang
    params['tl'] = tolang
    params['text'] = text

    req = requests.get(__base_url, params=params, headers=__headers)
    if req.status_code == 200:
        transtext = re.search(matcher, req.text)
        print(transtext)
        if transtext:
            return transtext.group(1)

if __name__ == "__main__":
    t1 = do_trans("Hello World", "ar-BH", "en")
#     print(t1.encode('utf-8'))
#     t2 = do_trans(t1, "en", "ar-BH")
#     print(t2.encode('utf-8'))
```
